## 人工データ実験

```python
def data1(view=True):
    ipv = np.ones(2)/2 #初期確率
    tpm = np.array([[0.95,0.05],
                    [0.05,0.95]]) #遷移確率
    tens = np.array([[6,],
                     [1,]]) #強度
    seed = 2022
    N = 10000 #データの個数
    phm = Poisson_Hidden_Markov_generator(ipv,tpm,tens)
    xdata = phm.generate(N,seed)
    if view:
        phm.view_data()
    return xdata
```

```python
def data2(view=True):
    ipv = np.ones(3)/3
    tpm = np.array([[0.7,0.2,0.1],
                    [0.3,0.4,0.3],
                    [0.2,0.3,0.5]])
    tens = np.array([[16,1,],
                     [9,9,],
                     [1,16]])
    seed = 2022
    N = 10000
    phm = Poisson_Hidden_Markov_generator(ipv,tpm,tens)
    xdata = phm.generate(N,seed)
    if view:
        phm.view_data()
    return xdata
```

```python
xdata = data2()
phmm = Poisson_Hidden_Markov(xdata,3)
```

## 完全分解変分推論(completely factorized variational inference)
